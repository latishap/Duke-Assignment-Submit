```python
import pprint

pdf_list = []

for a in range(1,26):
    s = "Desktop\Duke\Duke-Assignment\Text Files\PDM Financial "+ str(a) + ".text"
    
    pdf_list.append(s)
    
def details(path):
    
    import re
    import json
    import pprint
    import os



    pdm_details = {"Date" : None,
                  "Carrier Name" : None,
                  "Carrier MC Number" : None,
                  "NOA Company Name": "PDM Financial LLC",
                  "LOR Company Name": None}


    #cleaning
    def data_clean(text):
        try:
            strip_po = [i.strip() for i in text.split('\\n')]
            remove_space = [i for i in strip_po if i not in [' ', '']]
            remove_space_between = [i.split('  ') for i in remove_space]
            cleaned_data = list()
            count = 0
            for i in remove_space_between:
                cleaned_data.append([])
                for j in i:
                    if j not in ['']:
                        cleaned_data[count].append(j.strip())
                count += 1

            return cleaned_data

        except Exception as e:
            print("kvt data clean exception")
            print(e)
            return None

    f = open(path, "r")
    tmp = f.read()

    tmp2 = data_clean(tmp)


    #for date
    def get_date(origtmp):

        regex_lc = r'[a-zA-Z]{3,8} \d{2}, \d{4}'
        compiled_re = re.compile(regex_lc)


        for line in origtmp:
            for wordstr in line:
                lc = compiled_re.findall(wordstr)
                if len(lc) > 0:
                    return(lc[0])
        return('')


    pdm_details["Date"] = get_date(tmp2)

    x = [get_date(tmp2)]

    i = tmp2.index(x)

    pdm_details['Carrier Name'] = tmp2[i+1]

    #for MC no
    def get_mc_no(origtmp):

        regex_lc = r'MC \d{8}'
        compiled_re = re.compile(regex_lc)

        for line in origtmp:
            for wordstr in line:
                lc = compiled_re.findall(wordstr)

            if len(lc) > 0:
                return(lc[0])

        return ('')

    pdm_details['Carrier MC Number'] = get_mc_no(tmp2)

    #for letter of release
    def get_lor(origtmp):

        regex_lc = r'\D{100} RE:'
        compiled_re = re.compile(regex_lc)

        for line in origtmp:
            for wordstr in line:
                lc = compiled_re.findall(wordstr)

            if len(lc) > 0:
                return(lc[0])

        return ('')

    pdm_details['LOR Company Name'] = get_lor(tmp2)

    #for notice of assignment
    def get_noa(origtmp):

        regex_lc = r'\D{5,100} RE:'
        compiled_re = re.compile(regex_lc)

        for line in origtmp:
            for wordstr in line:
                lc = compiled_re.findall(wordstr)

            if len(lc) > 0:
                return(lc[0])

        return ('')

    pdm_details['NOA Company Name'] = tmp2[i+1]


    pprint.pprint(pdm_details)
    
    return 0

for l in pdf_list :
    
    try :
        
        details(l)
        
    except ValueError :
        
        print("Didn't Find Anything\n")
```
